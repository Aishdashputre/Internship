```python
# QUE 2 Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.
import re
import pandas as pd
url = 'http://www.imdb.com/chart/top'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
movies = soup.select('td.titleColumn')
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value')
        for b in soup.select('td.posterColumn span[name=ir]')]
 
 
 
 
# create a empty list for storing
# movie information
list = []
 
# Iterating over movies to extract
# each movie's details
for index in range(0, 100):
     
    # Separating movie into: 
    # 'title', 'year'
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    place = movie[:len(str(index))-(len(movie))]
    data = {
            "movie_title": movie_title,
            "rating": ratings[index],
            "year": year,
            }
    list.append(data)
 
# printing movie details with its rating.
for movie in list:
    print(movie['movie_title'], '('+movie['year'] +
        ') -', movie['rating'])

df = pd.DataFrame(list)
df.to_csv('imdb_top_100_movies.csv',index=False)
```

```python
# QUE 5 Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
 # a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
 # b) Top 10 ODI Batsmen along with the records of their team and rating.
 # c) Top 10 ODI bowlers along with the records of their team and rating.


from bs4 import BeautifulSoup
import requests
import pandas as pd
info=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup=BeautifulSoup(info.content)
team=soup.find_all('tr',class_=('rankings-block__banner','table-body'))
top10=team[0:10]
data = {'Team_Name':[],'Matches': [],'Points': [],'Rating':[]}
for i in top10:
    pnt=i.find_all('td',recursive=True)
    data['Team_Name'].append(i.find('span',class_='u-hide-phablet').text)
    data['Matches'].append(pnt[2].text)
    data['Points'].append(pnt[3].text)
    data['Rating'].append(pnt[4].text.replace('\n',''))
MenTeam=pd.DataFrame(data,index=range(1,11))
MenTeam

```

```python
# QUE 7 Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :
#i) Headline
#ii) Time
#iii) News Link


req=requests.get('https://www.cnbc.com/world/?region=world')
soup=BeautifulSoup(req.content)
headline = []
for i in soup.find_all('a',class_="LatestNews-headline"):
    headline.append(i.text)
headline
```

```python
time = []
for i in soup.find_all('time',class_="LatestNews-timestamp"):
    time.append(i.text)
time
```

```python
news_link = []
for i in soup.find_all('a',class_="LatestNews-headline"):
    news_link.append(i.get('href'))
news_link
```

```python
import pandas as pd
df = pd.DataFrame({'Headline' : headline, 'Time' : time, 'NewsLink' : news_link})
df
```

```python
# QUE 9 Write a python program to scrape mentioned details from dineout.co.in :
#i) Restaurant name
#ii) Cuisine
#iii) Location
#iv) Ratings
#v) Image URL



req=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup=BeautifulSoup(req.content)
restaurant_name = []
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    restaurant_name.append(i.text)
restaurant_name
```

```python

location = []
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location
```

```python
sta = []
for i in soup.find_all('span',class_="double-line-ellipsis"):
    sta.append(i.text.split('|')[1])
sta
```

```python
rat = []
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rat.append(i.text)
rat
```

```python
images = []
for i in soup.find_all('img',class_="no-img"):
    images.append(i.get('data-src'))
images
```

```python
df = pd.DataFrame({'Restaurant name':restaurant_name, 'Cuisine': sta, 'Location': location, 'Rating': rat, 'Image_URL':images})
df
```
